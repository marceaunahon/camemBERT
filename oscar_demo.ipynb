# Initialization

```python
from oscar import Oscar
oscar = Oscar()
```

# Tokenization example

```python
# Tokenize a sentence
sentence = "On parle francais ici! comme anticonstitutionnellement par exemple mon pote"
tokens = oscar.tokenize_text(sentence)
print(tokens)
tokens_ids = oscar.tokens_to_ids(tokens)
print(tokens_ids)
```

# Different Oscar methods output

```python
print(oscar[32])
print(oscar.get_item(32))
print(oscar.get_raw_text(32))
print("Oscar size:",len(oscar))
print(oscar.get_random_sample())
```

# Get vocab

```python
# get the vocab
print(oscar.get_vocab())
# get the vocab size
print(oscar.get_vocab_size())
```

# Searching method

```python
example_with_keyword = oscar.search_by_keyword("John")

print(f"number of examples with keyword: {len(example_with_keyword)}")
for example in example_with_keyword:
    print(example)
```

# CamemBERT real Tokenization

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("camembert-base")

tokenized = tokenizer.tokenize("Je suis anticonstitutionnellement pomme de terre.")
print(tokenized)

# Get CamemBERT vocab
camembert_vocab = tokenizer.get_vocab()

# Sort tokens based on their indices
# sorted_tokens = sorted(camembert_vocab.items(), key=lambda item: item[1])

# Print the size of CamemBERT vocab
print(len(camembert_vocab))

# add start and end tokens
tokenized = ["<s>"] + tokenized + ["</s>"]

# convert tokens to ids
tokens_ids = tokenizer.convert_tokens_to_ids(tokenized)
print(tokens_ids)
```
